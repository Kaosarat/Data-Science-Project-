### QUESTION 1 - Student Score Tracker 

```python
# put the scores in a list

student_score = [65, 78, 90, 55, 88, 72]
print(student_score)
```

```python
# print the highest score in the list

print(max(student_score))
```

```python
# print the lowest score in the list

failed_score = min(student_score)
print(failed_score)
```

```python
# calculate and print the average score

total_score = sum(student_score)
num_of_score = len(student_score)
avg_score = total_score/num_of_score
print(avg_score)
```

```python
# add a new test score of 80 to the list
student_score.append(80)
print(student_score)
```

```python
# remove the failed score from the list

student_score.remove(55) 
```

```python
print(student_score)
```

### QUESTION 2 - Shopping Cart System

```python
# put the items in list

cart = ["rice", "beans", "oil", "tomatoes"]
print(cart)
```

```python
# add two new item of your choice to the list

cart.append("onion")
cart.append("salt")
print(cart)
```

```python
# remove one item that the customer no longer wants 

cart.remove("oil")
print(cart)
```

```python
# print the total number of items currently in the cart

print(len(cart))
```

### QUESTION 3 - Daily Temperature Record

```python
# the temperatures for a week

temperature = (28, 30, 31, 29, 27)  #weekdays temperature from monday
print(temperature)
```
